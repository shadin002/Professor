# **Numpy Operations**

```python
n1=np.array([[1,2,3],[4,7,6]])
n1.shape
```

```python
n1.shape=(3,2)
n1
```

```python
#vstack()

n1=np.array([1,2,3])
n2=np.array([4,5,6])
np.vstack((n1,n2))
```

```python
n1=np.array([1,2,3])
n2=np.array([4,5,6])
np.vstack((n2,n1))
```

```python
#hstack()
n1=np.array([1,2,3])
n2=np.array([4,5,6])
np.hstack((n1,n2))
```

```python
n1=np.array([1,2,3])
n2=np.array([4,5,6])
np.hstack((n2,n1))
```

```python
#column_stack()
n1=np.array([1,2,3])
n2=np.array([4,5,6])
np.column_stack((n1,n2))
```

```python
n1=np.array([1,2,3])
n2=np.array([4,5,6])
np.column_stack((n2,n1))
```

```python
#inttersect1d()
n1=np.array([10,20,30,40,50,60])
n2=np.array([40,50,60,70,80,90,100,70])
np.intersect1d(n1,n2)
```

```python
n1=np.array([10,20,30,40,50,60])
n2=np.array([40,50,60,70,80,90,100,70])
np.intersect1d(n2,n1)
```

```python
#setdiff1d
n1=np.array([10,20,30,40,50,60])
n2=np.array([40,50,60,70,80,90,100,70])
np.setdiff1d(n1,n2)
```

```python
n1=np.array([10,20,30,40,50,60])
n2=np.array([40,50,60,70,80,90,100,70])
np.setdiff1d(n2,n1)
```

# **Numpy Mathematics**

```python
n1=np.array([10,20])
n2=np.array([30,40])

np.sum([n1,n2]) #sum of all elements
```

```python
np.sum([n1,n2],axis=0) #axis=0 is the sum based on column
```

```python
np.sum([n1,n2],axis=1) #axis=1 is the sum based on row
```

```python
#basic arithmetic operations
n1=np.array([10,20,30])
n1+5
```

```python
n1-5
```

```python
n1*2
```

```python
n1/5
```

```python
np.mean(n1)
```

```python
n2=np.array([10,20,30,40])
np.median(n2)
```

```python
np.std(n1)
```
