#**Line Plot**

**Plottting 2 point**

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1,11)
x

y=x*2
y

plt.plot(x,y)
plt.show()
```

Adding label and changing color


```python
plt.plot(x,y)
plt.title("First plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
plt.plot(x,y,color="orange",linestyle=":",linewidth=4)
plt.title("First plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
x=np.arange(1,11)
y1=x*2
y2=x*3
plt.plot(x,y1,color="blue",linestyle=":",linewidth=3)
plt.plot(x,y2,color="orange",linestyle="-",linewidth=4)
plt.title("First plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid("True")
plt.show()
```

Plotting into 2 plot

```python
# x
# y1
# y2
plt.subplot(2,1,1)
plt.plot(x,y1,color="orange",linestyle=":",linewidth=3)

plt.subplot(2,1,2)
plt.plot(x,y2,color="green",linestyle="-",linewidth=5)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="orange",linestyle=":",linewidth=3)

plt.subplot(1,2,2)
plt.plot(x,y2,color="green",linestyle="-",linewidth=5)

plt.show()
```

#**Bar Plot**

```python
student={"bob":87,"matt":56,"sam":27}
student

x=list(student.keys())
y=list(student.values())
plt.bar(x,y,color="cyan")
plt.title("Markks of student",color="blue")
plt.xlabel("Name")
plt.ylabel("Mark")
plt.grid(True)
plt.show()
```

```python
student={"bob":87,"matt":56,"sam":27}
student

x=list(student.keys())
y=list(student.values())
plt.barh(x,y,color="yellow")
plt.title("Markks of student",color="blue")
plt.xlabel("Mark")
plt.ylabel("Name")
plt.grid(True)
plt.show()
```

#**Scatter Plot**

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,3,1,5,7,8,9,6]

plt.scatter(x,y,color="red",marker="*",s=200)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y1=[2,4,3,1,5,7,8,9,6]
y2=[3,5,7,8,6,9,2,1,4]

plt.scatter(x,y1,color="red",marker="*",s=200)
plt.scatter(x,y2,color="yellow",marker="o",s=100)
plt.show()
```

In 2 different plot

```python
x=[1,2,3,4,5,6,7,8,9]
y1=[2,4,3,1,5,7,8,9,6]
y2=[3,5,7,8,6,9,2,1,4]

plt.subplot(1,2,1)
plt.scatter(x,y1,color="red",marker="*",s=200)
plt.subplot(1,2,2)
plt.scatter(x,y2,color="yellow",marker="o",s=100)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y1=[2,4,3,1,5,7,8,9,6]
y2=[3,5,7,8,6,9,2,1,4]

plt.subplot(2,1,1)
plt.scatter(x,y1,color="red",marker="*",s=200)
plt.subplot(2,1,2)
plt.scatter(x,y2,color="yellow",marker="o",s=100)
plt.show()
```

Working with dataframe

```python
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)
iris['species'] = data.target
iris.head()
plt.hist(iris["petal length (cm)"],color="green",bins=30)
plt.show()
```

```python
fruit=["apple","mango","orange","litchi"]

quantity=[60,40,50,25]

plt.pie(quantity,labels=fruit)
plt.show()
```
